## Tokenization

This function will assign tokens (numbers) to each word to improve the performance of other functions.
First we will get data from a webpage

```python
from bs4 import BeautifulSoup 
import urllib.request
 
response = urllib.request.urlopen('http://php.net/') 
html = response.read() 
soup = BeautifulSoup(html,"html5lib") 
text = soup.get_text(strip=True)
 
print (text)
```

Now that we have our cleaned text, we can convert this cleaned text into tokens

```python
tokens = [t for t in text.split()]
 
print (tokens)
```

## Word frequency

With the previous tokens you can easily get the word frequency with the following code:

```python
freq = nltk.FreqDist(tokens)
 
for key,val in freq.items():
 
    print (str(key) + ':' + str(val))
```

```python
freq.plot(20, cumulative=False)
```

## Word and Sentence Tokenization with NLTK

Word tokenitazion through NLTK is an improvement compare to the basic function Split
NLTK provide a function that can recognize in a better way each word in a sentence.

```python
from nltk.tokenize import word_tokenize
 
mytext = "Hello Mr. Adam, how are you? I hope everything is going well. Today is a good day, see you dude."
print(word_tokenize(mytext))
```

Words like "Mr." can be easly recognize thanks to NLTK. Also it recognize punctuation as a word.

NLTK can also recognize sentenses in a text. Let see how it works.

```python
from nltk.tokenize import sent_tokenize
 
mytext = "Hello Adam, how are you? I hope everything is going well. Today is a good day, see you dude." 
print(sent_tokenize(mytext))
```

You can think that this is easy and you don't need NLTK to get the same results because you can divide sentenses using regular expressions given that every sentense is separated by a dot and a space. Let's see something different.

```python
from nltk.tokenize import sent_tokenize
 
mytext = "Hello Mr. Adam, how are you? I hope everything is going well. Today is a good day, see you dude." 
print(sent_tokenize(mytext))
```

In this case NLTK recognize the first sentense until the question mark, considering the word "Mr." as part of the sentense

## Tokenize text in other languages

NLTK can tokenize text in different languages. Please refer to NLTK documentation to find all available languages.

```python
from nltk.tokenize import sent_tokenize
 
mytext = "Bonjour M. Adam, comment allez-vous? J'espère que tout va bien. Aujourd'hui est un bon jour." 
print(sent_tokenize(mytext,"french"))
```

## Definitions, Synonyms and Antonyms from WordNet

One of the packages you can download from NLTK.download() is WordNet.
WordNet is a database created for NLP. It has nynonyms, antonyms and definitions.

```python
from nltk.corpus import wordnet
 
syn = wordnet.synsets("pain") 
print(syn[0].definition()) 
print(syn[0].examples())
```

For nynonyms you can use a code like this

```python
from nltk.corpus import wordnet
 
synonyms = [] 
for syn in wordnet.synsets('Computer'): 
    for lemma in syn.lemmas(): 
        synonyms.append(lemma.name())
 
print(synonyms)
```

For antonyms you can do the following:

```python
from nltk.corpus import wordnet
 
antonyms = [] 
for syn in wordnet.synsets("small"): 
    for l in syn.lemmas(): 
        if l.antonyms(): 
            antonyms.append(l.antonyms()[0].name())
 
print(antonyms)
```
